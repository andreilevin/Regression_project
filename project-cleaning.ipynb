```python
def lowercasestrip(string):
    try:
        string = string.replace('-','').replace('_','').replace(' ','').replace('.','').replace('\'','').lower()
    except:
        pass
    return string

def dollarstoint(string):
    string = string.replace('$','').replace(',','') 
    try:
        return int(string)
    except:
        return string
```

```python
def transform_sal(df):
    df['Player'] = df['1'].apply(lowercasestrip)
    df['Salary'] = df['2'].apply(dollarstoint)
    df = df.drop(columns= ['Unnamed: 0','1','2'])
#   df.drop(index=0).reset_index(drop=True)
    df = df.set_index('Player',drop=True)
    return df
```

```python
dfsal2016_ = transform_sal(dfsal2016)
dfsal2017_ = transform_sal(dfsal2017)    
dfsal2018_ = transform_sal(dfsal2018)
dfsal2019_ = transform_sal(dfsal2019)
dfsal2020_ = transform_sal(dfsal2020)
```

```python
def heightinfeet(height):
    try:
        height_ft  = float(height.split('-')[0]) + float(height.split('-')[1])/12
        return height_ft
    except:
        return np.nan
    
    
minsalary = {2016:   875000, 2017:  1313000, 2018:   1339000,  2019:    1446000,   2020:   1446000}
salarycap = {2016: 94143000, 2017: 99093000, 2018: 101869000,  2019:  109140000,   2020: 117000000}    
    
def dfcleaned(dfyear, prevyear):
    df = dfyear.copy()
    dropcols = ['weight', 'height', 'Age', 'Prev_Sal', 'Next_Sal', 'FG%', 'USG%', 'TS%', 'G']
    df = df.dropna(subset=dropcols)
    df = df[df.G != 0]
    
    df['3P%'].fillna(0, inplace=True)   # fill with median
    df['FT%'].fillna(df['FT%'].median(), inplace=True)   # fill with median
    
    df['height_ft'] = df['height'].apply(heightinfeet)

    cols = ['PTS','AST','Age','G','GS','MP','3P','FT','ORB', 'TRB', 'AST', 'STL', 'BLK',
            'TOV', 'PTS','3P%','FT%','FG%']
    for col in cols:
        df[col] = df[col].apply(lambda s: pd.to_numeric(s, errors='coerce'))
        
    df = df[df.Prev_Sal > minsalary[prevyear]] 
    df = df[df.Next_Sal > minsalary[prevyear+1]]   
    
    df['Prev_Sal'] = df['Prev_Sal']*salarycap[2020]/salarycap[prevyear]/1000000
    df['Next_Sal'] = df['Next_Sal']*salarycap[2020]/salarycap[prevyear+1]/1000000
     
    return df

dfcleaned(dftest,2018).describe()
```
